```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/datasalesfinal/data_sales .csv')
print(df)
```

```python
#This loads the sales CSV into pandas and prints the data.
```

```python
slice_north = df[df['Region'] == 'North']
print(slice_north)
```

```python
#This filters the dataset to only include records from the North region
```

```python
dice_north_electronics = df[(df['Region'] == 'North') & (df['Category'] == 'Electronics')]
print(dice_north_electronics)
```

```python
#This filters for North region AND Electronics category.
```

```python
rollup_total_sales = df.groupby('Region')['Sales'].sum().reset_index()
print(rollup_total_sales)
```

```python
#Groups sales data by Region and sums the Sales column.
```

```python
drilldown_sales = df.groupby(['Region', 'Month'])['Sales'].sum().reset_index()
print(drilldown_sales)
```

```python
#Groups by Region and Month for more detailed aggregation.
```

```python
pivot_table = pd.pivot_table(df, values='Sales', index='Region', columns='Month', aggfunc='sum', margins=True, margins_name='Total')
print(pivot_table)
```

```python
#Creates a pivot table showing total sales by Region and Month, with grand totals.
```
